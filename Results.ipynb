```python
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
```

# Training Curves

```python
loss_duel = 'dueling-logs/losses.csv'
loss_pr = 'prioritized-logs/losses.csv'
rewards_duel = 'dueling-logs/rewards.csv'
rewards_rb = 'rainbow-logs/rewards.csv'
rewards_pr = 'prioritized-logs/rewards.csv'
```

## Dueling

```python
loss_df1 = pd.read_csv(loss_duel)
loss_df2 = pd.read_csv(loss_pr)
plt.plot(loss_df1['index'],loss_df1['loss'],label = 'Dueling')
plt.plot(loss_df2['index'],loss_df2['loss'],label = 'Prioritized Experience Replay')
plt.xlabel('Iterations')
plt.ylabel('Loss')
plt.title('Loss during training')
plt.legend()
plt.show()
```

## Rewards plot

```python
rewards_df1 = pd.read_csv(rewards_duel)
rewards_df2 = pd.read_csv(rewards_pr)
plt.plot(rewards_df1['episode'],rewards_df1['reward'],label = 'Dueling')
plt.plot(rewards_df2['episode'],rewards_df2['reward'],label = 'Prioritized')
plt.xlabel('Episodes')
plt.ylabel('Rewards')
plt.title('Rewards during training')
plt.legend()
plt.show()
```

```python
rewards_rb = 'rainbow-logs/rewards_comb.csv'
loss_rb = 'rainbow-logs/losses_comb.csv'
rewards_rb1 = 'rainbow_imp-logs/rewards.csv'
loss_rb1 = 'rainbow_imp-logs/losses.csv'

loss_df1 = pd.read_csv(loss_rb)
loss_df2 = pd.read_csv(loss_rb1)

rewards_df1 = pd.read_csv(rewards_rb)
rewards_df2 = pd.read_csv(rewards_rb1)
```

```python
plt.plot(rewards_df1['episode'],rewards_df1['rewards'],label = 'Rainbow (RELU)')
plt.plot(rewards_df2['episode'],rewards_df2['rewards'],label = 'Rainbow (SELU)')
plt.xlabel('Episodes')
plt.ylabel('Rewards')
plt.title('Rewards during training')
plt.legend()
plt.show()
```
